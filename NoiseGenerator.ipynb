```python
'''
Adds noise to and saves lena image.
'''

# load skimage function
from skimage.io import imread, imsave, imshow
from skimage.util import random_noise

# load original lena image
lena_img = imread('lena_imgs/lena.jpg')

# add noise and save image
imsave('lena_imgs/lena_gaussian.jpg', random_noise(lena_img, mode='gaussian'))
imsave('lena_imgs/lena_poisson.jpg', random_noise(lena_img, mode='poisson'))
imsave('lena_imgs/lena_salt.jpg', random_noise(lena_img, mode='salt'))
imsave('lena_imgs/lena_pepper.jpg', random_noise(lena_img, mode='pepper'))
imsave('lena_imgs/lena_speckle.jpg', random_noise(lena_img, mode='speckle'))
```
